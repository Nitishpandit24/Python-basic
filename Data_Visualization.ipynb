1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, show
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.




```python
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook


n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(5, 50, size=n)
colors = np.random.choice(['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD'], size=n)


data = pd.DataFrame({'x': x,'y': y,'sizes': sizes,'colors': colors})
source = ColumnDataSource(data)
output_notebook()
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', tools="pan,box_zoom,reset")
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.layouts import layout
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap Example", x_axis_label='x', y_axis_label='y', tooltips=[("x", "$x"), ("y", "$y"), ("value", "@image")])

color_mapper = LinearColorMapper(palette=Viridis256, low=np.min(data_heatmap), high=np.max(data_heatmap))

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title='Value')
p.add_layout(color_bar, 'right')

output_notebook()

show(p)
```
