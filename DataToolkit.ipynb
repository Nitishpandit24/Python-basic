3. Explain the following terms:

a. The difference in np.array, np.asarray and np.asanyarray

b. The difference between Deep copy and shallow copy


Ans:-
a}.  np.array will always create a new array unless copy=False is specified. np.asarray and np.asanyarray will not copy if the input is already an array but np.asanyarray preserves subclasses of ndarray, whereas np.asarray and np.array do not.

b}. : A shallow copy will create a new object, but it does not create copies of objects that are contained within the original object,Instead, it only copies the references to the original objects.

Where as, A deep copy creates a new object and recursively copies all objects found in the original object, ensuring that no references to objects in the original object are kept.

15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import scipy.stats as stats

data=np.random.normal(loc=0,scale=10,size=1000)
plt.hist(data,bins=30, density=True, alpha=0.6)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel("Value")
plt.ylabel("Frequency/Probability")
plt.title("Histogram with PDF Overlay")
plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

data1=np.random.rand(100)
data2=np.random.rand(100)
df=pd.DataFrame({"data1":data1,"data2":data2})

sns.relplot(data=df,x="data1",y="data2",)
conditions = [(df['data1'] >= 0) & (df['data2'] >= 0),
              (df['data1'] < 0) & (df['data2'] >= 0),
              (df['data1'] < 0) & (df['data2'] < 0),
              (df['data1'] >= 0) & (df['data2'] < 0)
              ]
choices = ['First', 'Second', 'Third', 'Fourth']
df['Quadrant'] = np.select(conditions, choices, default='None')

plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='data1', y='data2', hue='Quadrant', palette='bright')

plt.legend(title="Quadrant")
plt.xlabel("data1")
plt.ylabel("data2")
plt.title("Quadrant-wise Scatter Plot")
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'8

```python
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, show
bokeh.io.output_notebook()

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
p = figure(title='Sine Wave Function', x_axis_label='x (radians)', y_axis_label='sin(x)')
p.line(x, y, legend_label='sin(x)', line_width=2)

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
from bokeh.models import ColumnDataSource, HoverTool

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))
colors = [f'#{int(255 - (value * 2.55)):02x}00ff' for value in values]
source = ColumnDataSource(data=dict(categories=categories, values=values, colors=colors))
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values',
           tools="")
bars = p.vbar(x='categories', top='values', source=source, width=0.5,
              line_color='white', fill_color='colors')
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'8

```python
import numpy as np
import plotly.graph_objects as go

x = np.arange(10)
y = np.random.rand(10) * 100
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Random Data'))
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

fig.show()
```
